```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

save_option = True
figsavepath = "../Figures/FigureS3_solubility.svg"

path = "../Data/FigureS3_solubility/230628 Solubility-test.xlsx"
df = pd.read_excel(path)

x = df['T'].to_list()
y = df['avg'].to_list()
x = x[:7]
y = y[:7]

#fitting
coefficients = np.polyfit(x,y,2)
polynomial = np.poly1d(coefficients)
x_fit = np.linspace(18,85,100)
y_fit = polynomial(x_fit)

### Plot ###
plt.rcParams["font.size"] = 15
fig = plt.figure(figsize=(5,5))
color = "#474D6B" # blue gray

plt.scatter(x,y, label='Experiment', color='#474D6B80',ec=color)
plt.plot(x_fit, y_fit, label='Fitting', color=color)
plt.xlabel('Temperature (â„ƒ)')
plt.ylabel('Solubility (wt.%)')
plt.legend(frameon=False)
# plt.text(19, 28, '$\it{y} = 0.00166\it{x}^{2} - 0.54452\it{x} + 61.155$', ha='left', va='bottom',color=color,fontsize=15)
plt.text(19, 28, '$\it{S} = 0.00166\it{T}^{2}$', ha='left', va='bottom',color=color,fontsize=15)
plt.text(29, 26, '$- 0.55445\it{T} + 61.155$', ha='left', va='bottom',color=color,fontsize=15)

print(coefficients)

if save_option:
    plt.savefig(figsavepath, dpi=1200, bbox_inches='tight')
```
