```python
import pandas as pd
import matplotlib.pyplot as plt

# figure save option
save_option = True
figsavepath = "../Figures/FigureS15_CsPbBr3.svg"

path = "../Data/FigureS15_CsPbBr3/2023-11-22 data.xlsx"
df = pd.read_excel(path, sheet_name='Data')
X = df['Time (h)'].to_list()
Y = df['Smth_G (mm/h)'].to_list()

X_points = [10,20,30,40] # check points (hrs)
Y_points = []
i = 0
for x, y in zip(X,Y):
    if x > X_points[i]:
        Y_points.append(y)
        i += 1
        if i == 4:
            break
        
### Figure Plot ###
fig = plt.figure(figsize=(12,5))
plt.rcParams["font.size"] = 15

plt.xlabel('Time (h)')
plt.ylabel('Growth rate (mm/h)')
plt.plot([0,100],[0.1,0.1],c="#474D6B80",linestyle='dashed',label='Set value (0.1)')
plt.plot(X,Y,color="#474D6B")
plt.scatter(X_points,Y_points,color="#474D6B", s=80)
plt.xlim(0,42)
plt.ylim(0,0.35)

if save_option:
    plt.savefig(figsavepath, dpi=1200, bbox_inches='tight',transparent=True)
    
plt.show()
```
