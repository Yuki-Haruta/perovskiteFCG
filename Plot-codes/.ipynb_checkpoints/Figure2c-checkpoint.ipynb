```python
import pandas as pd
import glob
import os
import openpyxl
import matplotlib.pyplot as plt

dir_path = "../Data/FCG-excels"
files = glob.glob(f'{dir_path}/*.xlsx')
files.sort()

X, Y = [], []


for file in files:
    if "$" in file:
        pass
    else:
        try:
            # glab the data
            df = pd.read_excel(file, sheet_name ='Data')

            # Read the second and third sheet
            wb = openpyxl.load_workbook(file)
            ws_P = wb["Parameters"]
            ws_I = wb["Information"]

            # ID
            ID = os.path.basename(file)[:-5]

            # SV
            SV = ws_P['I6'].value # Set growth rate (mm/h)

            # sPV calculation
            Smth_G = df['Smth_G (mm/h)'].to_list()
            last_G = Smth_G[-2169:] #cut the last 3 hour
            sPV = sum(last_G)/len(last_G) # average growth rate at the last 3 hours (mm/h)
            
            if isinstance(SV, float):
                X.append(SV)
                Y.append(sPV)

            print(f'ID: {ID}, SV: {SV:.3f}, sPV: {sPV:.2f}')
        except Exception as e:
            print(e)
plt.scatter(X,Y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

# read the data
# path = "../Data/Figure2/SV-sPV_240227.xlsx"

# df = pd.read_excel(path)

# X = df['SV'].to_list()
# Y = df['PV'].to_list()
Z = [m-n for m,n in zip(Y,X)]

# Figure 1
plt.rcParams["font.size"] = 17
fig = plt.figure(figsize=(5,5))
color = "#474D6B" # dark blue

# Fitted curve
X_fit = np.arange(0,0.71,0.01)
Y_fit = X_fit
# calculate R-squared value
Y_fit2 = [n for n in X]
R2 = r2_score(Y, Y_fit2)

print(f'R2: {R2}')

# fit
plt.plot(X_fit, Y_fit, linestyle='dashed', color='#474D6B')

plt.fill_between(X_fit, [n+0.1 for n in Y_fit], [n-0.1 for n in Y_fit], facecolor=color, alpha=0.2)
plt.fill_between(X_fit, [n+0.05 for n in Y_fit], [n-0.05 for n in Y_fit], facecolor=color, alpha=0.2)

# main plot
plt.scatter(X,Y, s=30, color='#474D6B',ec='#474D6B', linewidths=1)

# design
plt.xlabel('SV (mm/h)')
plt.ylabel('stabilized PV (mm/h)')
plt.xlim(0,0.7)
plt.ylim(0,0.79)
plt.xticks(np.arange(0, 0.8, 0.1))
plt.yticks(np.arange(0, 0.8, 0.1))
plt.text(0.05,0.68, '$\it{y} = \it{x}$', ha='left', va='bottom',color=color,fontsize=17)
plt.text(0.05,0.61, '$(\it{R}^{2} = 0.75)$', ha='left', va='bottom',color=color,fontsize=17)

figsavepath = "../Plots"
plt.savefig(f'{figsavepath}/Figure2c.svg', dpi=1200, bbox_inches='tight', transparent=True)

plt.show()
```
