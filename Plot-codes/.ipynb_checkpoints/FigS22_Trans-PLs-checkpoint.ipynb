```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
from scipy.special import wofz, gamma
from sklearn.metrics import r2_score
from scipy.stats import pearsonr

save_option = False
figsavepath = "../Figures/FigureS16.svg"

# Transmittance
pathT = '../Data/Figure5/230811-A.Sample.csv' # high grade

# Single-Photon PL
pathSP = '../Data/Figure5/230811_A_SP_405nm_0.25_power_2.txt' # high grade

# Two-Photon PL
pathTP = '../Data/Figure5/230811_A_2P_800nm_0.4_power_inside.txt' # high grade

### Figure Plot ###
plt.rcParams["font.size"] = 15
fig, axes = plt.subplots(nrows=2, ncols=1, tight_layout=True, figsize=(8,8))

# Panel 1
df_T = pd.read_csv(pathT, skiprows=1)
x_T = df_T['nm'].to_list()
x_T = [1240/n for n in x_T] # nm to eV
y_T = df_T[' %T'].to_list()

axes[0].plot(x_T,y_T,color='#01ADC1',linewidth=2, label='Transmittance')
axes[0].set_xlim(1.95,2.5)
axes[0].set_ylim(-5,70)
axes[0].set_xticks([])
axes[0].set_ylabel('Transmittance (%)')
# axes[0].legend(frameon=False)

# PL
df_SP = pd.read_csv(pathSP, sep='\t')
df_TP = pd.read_csv(pathTP, sep='\t')

x = df_SP['Emission wavelength [nm]'].to_list()
x_SP = [1240/n for n in x] # nm to eV
x_SP = x_SP[:-1]
y = df_SP['Region 1'].to_list()
y_SP = [n/max(y) for n in y] # normalize
y_SP = y_SP[:-1]

x = df_TP['Emission wavelength [nm]'].to_list()
x_TP = [1240/n for n in x] # nm to eV
x_TP = x_TP[:-1]
y = df_TP['Region 1'].to_list()
y_TP = [n/max(y) for n in y] # normalize
y_TP = y_TP[:-1]


axes[1].plot(x_SP,y_SP,label='SPA-PL',color='red',marker='o',linewidth=2)
axes[1].plot(x_TP,y_TP,label='TPA-PL',color='green',marker='^',linewidth=2)
axes[1].set_xticks([2,2.1,2.2,2.3,2.4,2.5])
axes[1].set_xlim(1.95,2.5)
axes[1].set_yticks([])
axes[1].set_xlabel("Energy (eV)")
axes[1].set_ylabel("Intensity (a.u.)")
axes[1].legend(frameon=False)

if save_option:
    plt.savefig(figsavepath, dpi=1200, bbox_inches='tight')
```
