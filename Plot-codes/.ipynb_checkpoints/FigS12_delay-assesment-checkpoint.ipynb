```python
import pandas as pd
import matplotlib.pyplot as plt

save_option = True
figsavepath = "../Figures/FigureR_delay.svg"

path = "../Data/00_FCG-excels/230920_A_0.2mmh.xlsx"
df = pd.read_excel(path)

X = df['Time (h)'].to_list()
Y1 = df['e(t)'].to_list()
Y2 = df['Inf-rate (mL/h)'].to_list()

# plot
plt.rcParams["font.size"] = 17
# fig = plt.figure(figsize=(17/3*2,5))
fig, ax1 = plt.subplots(figsize=(8,6))
color1 = '#474D6B' # dark blue
color2 = '#01ADC1' # light blue

# plot e(t)
ax1.plot(X,Y1,c=color1)
ax1.set_ylim(-0.2,0.1)
ax1.set_xlabel('Time (h)')
ax1.set_ylabel('e(t) (mm/h)')

# plot infusion rate
ax2 = ax1.twinx()
ax2.plot(X,Y2,c=color2)
ax2.set_xlabel('Time (h)')
ax2.set_ylabel('Infusion rate (mL/h)', color=color2)
ax2.tick_params(axis='y', which='both',colors=color2)
ax2.spines['right'].set_color(color2)

plt.xlim(2,6)

if save_option:
    plt.savefig(figsavepath, dpi=1200, bbox_inches='tight')
```
