```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import numpy as np

# figure save option
save_option = False
figsavepath = "../Plots/Figure2b.svg"

# read data for panel b
paths = ["../Data/Figure2/2023-07-24 data.xlsx", # without feedback
         "../Data/Figure2/2023-09-20 data.xlsx" # with feedback SV0.2
        ]

df_list, X_list, Y_list = [], [], []
for path in paths:
    df = pd.read_excel(path)
    df = df[600:] # cut the first 10 mins
    df_list.append(df)
    X = df['Time (h)'].to_list()
    Y = df['Smth_G (mm/h)'].to_list()
    X_list.append(X)
    Y_list.append(Y)

# Plot
plt.rcParams["font.size"] = 17
fig = plt.figure(figsize=(15, 5))
gs = gridspec.GridSpec(1, 2, width_ratios=[7, 3])
color1 = "#474D6B" # dark blue
color2 = '#3498B0' # light blue

# Plot for panel b: control vs no-control
ax1 = plt.subplot(gs[0])
ax1.plot(X_list[0],Y_list[0], color=color1, linewidth=4, label="w/o feedback")
ax1.plot(X_list[1],Y_list[1], color=color2, label='with feedback control', linewidth=4,zorder=3)
ax1.plot([0,45],[0.2,0.2],linestyle='dashed', color=f"{color1}80", linewidth=3, zorder=4,label='Set value (0.2 mm/h)')
ax1.set_xlim(0,45)
ax1.set_ylim(0,2.1)
ax1.set_xticks(np.arange(0,50,5))
ax1.set_xlabel('Time (h)')
ax1.set_ylabel('Growth rate (mm/h)')
ax1.legend(frameon=False)

# 23
x = [0,1]
y2 = [2,2]
ax2 = plt.subplot(gs[1])
ax2.plot(x, y2, '-x')
ax2.set_xlabel('Steady-state error (mm/h)')
ax2.set_ylabel('Counts')

# Show and save the plot
plt.tight_layout()
if save_option:
    plt.savefig(figsavepath, dpi=1200, bbox_inches='tight', transparent=True)
plt.show()
```
