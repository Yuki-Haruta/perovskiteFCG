```python
import matplotlib.pyplot as plt
import glob
import pandas as pd
import os
import numpy as np
from scipy.optimize import curve_fit

info_path = f"{source_dir}/Summary.xlsx"

df_info = pd.read_excel(info_path)

csv_list = glob.glob(f'{source_dir}/*.csv')
csv_list.sort()

file_list = []
transit_time_list = []
sample_list = []
mobility_list = []
E_list_list = []
mobility_list_list = []

for csv in csv_list:
    # get the info
    filename = os.path.basename(csv)[:-4]
    fileinfo = filename.split('_')
    sample_id = f'{fileinfo[0]}_{fileinfo[1]}' #240220_AY3
    thickness = float(fileinfo[2][:-2]) # mm
    voltage = float(fileinfo[3][:-1]) # V
    
    print(f'\n{filename}')
    
    # grab approximation range and plot range from the df_info
    range1 = [float(df_info.loc[df_info['filename'] == filename, 'range1[0]']),
              float(df_info.loc[df_info['filename'] == filename, 'range1[1]'])
             ]
    range2 = [float(df_info.loc[df_info['filename'] == filename, 'range2[0]']),
              float(df_info.loc[df_info['filename'] == filename, 'range2[1]'])
             ]
    plot_range = [[float(df_info.loc[df_info['filename'] == filename, 'plot1[0]']),
              float(df_info.loc[df_info['filename'] == filename, 'plot1[1]'])],
                  [float(df_info.loc[df_info['filename'] == filename, 'plot2[0]']),
              float(df_info.loc[df_info['filename'] == filename, 'plot2[1]'])]
                 ]
    
    # read original data
    df = pd.read_csv(csv,header=None,names=['Time(s)','Y'])
    
    X = df['Time(s)'].to_list()
    Y = df['Y'].to_list()

    # define t = 0 as Y max
    x_offset = X[Y.index(max(Y))]
    X = [(n-x_offset)*1e6 for n in X] # s
    Y = [n/max(Y) for n in Y] # normalized
    cut_index = Y.index(1)
    X = X[cut_index:]
    Y = Y[cut_index:]
    
    # prepare log 10 for the fitting
    logX = [np.log10(x) for x in X[1:]]
    logY = [np.log10(y) for y in Y[1:]]
    
    # fitting for the first region
    for x in logX:
        if x < range1[0]:
            index11 = logX.index(x)
        if x > range1[1]:
            index12 = logX.index(x)
            break
    a1, b1 = np.polyfit(logX[index11:index12],logY[index11:index12],1)
    X_fit1 = np.arange(plot_range[0][0],plot_range[0][1],0.1)
    Y_fit1 = [a1*n+b1 for n in X_fit1]
    print(f'slope1: {a1:.3f}')
    
    # fitting for the second region
    for x in logX:
        if x < range2[0]:
            index21 = logX.index(x)
        if x > range2[1]:
            index22 = logX.index(x)
            break
    a2, b2 = np.polyfit(logX[index21:index22],logY[index21:index22],1)
    X_fit2 = np.arange(plot_range[1][0],plot_range[1][1],0.1)
    Y_fit2 = [a2*n+b2 for n in X_fit2]
    print(f'slope2: {a2:.3f}')
    
    t_transit = 10**((b2-b1)/(a1-a2))
    print(f'{t_transit:.2f} s')
    
    X_fit1_original = [10**n for n in X_fit1]
    Y_fit1_original = [10**n for n in Y_fit1]
    X_fit2_original = [10**n for n in X_fit2]
    Y_fit2_original = [10**n for n in Y_fit2]

    plt.plot(X,Y)
    plt.plot(X_fit1_original,Y_fit1_original,color='black')
    plt.plot(X_fit2_original,Y_fit2_original,color='black')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Time (s)')
    plt.ylabel('Output (a.u.)')
    
    plt.show()
    
    file_list.append(filename)
    transit_time_list.append(t_transit)
    
    if voltage == 100:
        ## 1/V time plot ##
        d = thickness
        voltages = [10,20,40,60,80,100]
        inverse_V = [1/n for n in voltages]
        t_transits = transit_time_list[-6:]

        plt.scatter(inverse_V,t_transits,c='orange')

        # linear regression
        coefficients = np.polyfit(inverse_V,t_transits, 1)
        slope, intercept = coefficients
        X_fit = np.arange(0,0.11,0.01)
        Y_fit = [slope*n+intercept for n in X_fit]
        plt.plot(X_fit,Y_fit,c='orange',linestyle='dashed')

        # prediction
        predicted_t_cutoffs = np.polyval(coefficients, inverse_V)

        # residual
        residuals = np.array(t_transits) - predicted_t_cutoffs

        # RSS
        RSS = np.sum(residuals**2)

        # SXX
        SXX = np.sum((np.array(inverse_V) - np.mean(inverse_V))**2)

        # Standard error
        SE_slope = np.sqrt(RSS / ((len(inverse_V) - 2) * SXX))

        mobility = d*d/slope*1e6
        mobility_max = (d+0.01)*(d+0.01)/(slope-SE_slope)*1e6
        mobility_min = (d-0.01)*(d-0.01)/(slope+SE_slope)*1e6
        print(f'\n{sample_id}')
        print(f': {mobility:.2f} cm2 V-1 s-1')
        print(f' min: {mobility_min:.2f} cm2 V-1 s-1')
        print(f' max: {mobility_max:.2f} cm2 V-1 s-1')
        sample_list.append(sample_id)
        mobility_list.append([mobility_min,mobility,mobility_max])
        plt.xlabel('1/V (/V)')
        plt.ylabel('transit time (s)')
        plt.show()
        
        ## E  plot
        E_list = [n/d for n in voltages] # V/cm
        mobility_list2 = [d*d/m/n*1e6 for m,n in zip(voltages,t_transits)]
        
        plt.scatter(E_list,mobility_list2)
        plt.xlabel('E (V/cm)')
        plt.ylabel('mobility (cm2/V/s)')
        plt.show()
```
