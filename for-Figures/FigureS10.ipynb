```python
import pandas as pd
import matplotlib.pyplot as plt

path = "../Data/FigureS10/2023-11-22 data.xlsx"
df = pd.read_excel(path, sheet_name='Data')
x = df['Time (h)'].to_list()
y = df['Smth_G (mm/h)'].to_list()

### Figure Plot ###
fig = plt.figure(figsize=(12,5))
plt.rcParams["font.size"] = 15

plt.xlabel('Time (h)')
plt.ylabel('Growth rate (mm/h)')
plt.plot(x,y,color="#474D6B")
plt.xlim(0,40)
plt.ylim(0,0.35)

figsavepath = "../Plots"
plt.savefig(f'{figsavepath}/FigureS10.svg', dpi=1200, bbox_inches='tight', transparent=True)

plt.show()
```
