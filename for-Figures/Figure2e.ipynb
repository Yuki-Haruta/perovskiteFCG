```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

# Read the data
file_path = '../Data/Figure2/SV-PV_230605.xlsx'
data = pd.read_excel(file_path)
X = data['SV'].to_list()
Y = data['PV'].to_list()

# Linear fitting
coefficients = np.polyfit(X, Y, 1) # linear fitting
slope = coefficients[0]
intercept = coefficients[1]
# Fitted curve
X_fit = np.arange(0,0.7,0.01)
Y_fit = [slope * n + intercept for n in X_fit]
# calculate R-squared value
Y_fit2 = [slope * n + intercept for n in X]
R2 = r2_score(Y, Y_fit2)

# plot
plt.rcParams["font.size"] = 17
fig = plt.figure(figsize=(17/3,5))
color = "#474D6B" # dark blue

plt.scatter(X,Y, s=80, color='#474D6B50',ec='#474D6B', linewidths=2)
plt.plot(X_fit, Y_fit, linestyle='dashed', color=color)
plt.xlabel('SV (mm/h)')
plt.ylabel('stabilized PV (mm/h)')
plt.xlim(-0.05,0.75)
plt.ylim(-0.05,0.79)
plt.xticks(np.arange(0, 0.8, 0.1))
plt.yticks(np.arange(0, 0.8, 0.1))
plt.text(0,0.65, '$\it{y} = 1.01\it{x}+0.03$', ha='left', va='bottom',color=color,fontsize=17)
plt.text(0,0.58, '$(\it{R}^{2} = 0.80)$', ha='left', va='bottom',color=color,fontsize=17)

figsavepath = "../Plots"
plt.savefig(f'{figsavepath}/Figure2e.svg', dpi=1200, bbox_inches='tight', transparent=True)
```
