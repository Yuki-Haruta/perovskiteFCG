```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

# get the files
excel_paths = [
    "../Data/FigureS8/2023-01-28 data.xlsx",
    "../Data/FigureS8/2023-09-20 data.xlsx",
    "../Data/FigureS8/2023-09-25 data.xlsx",
    "../Data/FigureS8/2023-02-14 data.xlsx",
    "../Data/FigureS8/2023-04-26 data.xlsx"
]

labels = ["0.1 mm/h", "0.2 mm/h", "0.3 mm/h","0.4 mm/h","0.5 mm/h"]

dfe_list, Xe_list, Ye_list = [], [], []
for path in excel_paths:
    df = pd.read_excel(path)
    # df = df[600:] # cut the first 10 mins
    dfe_list.append(df)
    Xe = df['Time (h)'].to_list()
    Ye = df['Smth_G (mm/h)'].to_list()
    Xe_list.append(Xe)
    Ye_list.append(Ye)

"""
plot all the data
"""
# plot
plt.rcParams["font.size"] = 17
fig = plt.figure(figsize=(17/3*2,5))
color = "#474D6B" # dark blue
colors = ['#01ADC1', '#3498B0', '#68839F', '#9B6E8E', '#CF597E'] # from blue to red

# representative control data
for j in range(5):
    i = 4-j
    plt.plot(Xe_list[i],Ye_list[i], color=colors[i], label=labels[i], linewidth=3)
plt.xlim(0,50)
plt.ylim(0,0.8)
plt.xlabel('Time (h)')
plt.ylabel('Growth rate (mm/h)')
plt.xticks(np.arange(0,55,5))
plt.legend(frameon=False)
             
figsavepath = "../Plots"
plt.savefig(f'{figsavepath}/FigureS8.svg', dpi=1200, bbox_inches='tight', transparent=True)
```
