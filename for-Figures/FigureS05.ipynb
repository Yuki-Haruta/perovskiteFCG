```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = "../Data/FigureS5/231023_evap-rate.xlsx"
df = pd.read_excel(path)
Y1 = df['rate'].to_list()

### Plot ###
plt.rcParams["font.size"] = 15
fig = plt.figure(figsize=(5,4))
color = "#474D6B" # blue gray

# Box plots
bp=plt.boxplot([Y1],
              patch_artist=True,  # enable detail change
              widths=0.5,  # box width
              zorder=1,
              whis=[0,100], #below 10% and upper 90% is fliers
              showfliers=False
              )

colors=['#474D6B'] # boxes, medians, fliers
colors2=['#474D6B','#474D6B'] # whiskers, caps

# box color
for b, c in zip(bp['boxes'], colors):
    b.set(color=c, linewidth=2)  # box frame
    b.set_facecolor('White') # box
# center line
for b, c in zip(bp['medians'], colors):
    b.set(color=c, linewidth=2)
# whisker
for b, c in zip(bp['whiskers'], colors2):
    b.set(color=c, linewidth=2)
# cap
for b, c in zip(bp['caps'], colors2):
    b.set(color=c, linewidth=2)
# outsider
for b, c in zip(bp['fliers'], colors):
    b.set(markeredgecolor=c, markeredgewidth=2)

mean, width = 1, 0.2
X1 = np.random.uniform(mean-width, mean+width, len(Y1))
X1 -= np.mean(X1)
X1 += mean # mean value


plt.xlabel('')
plt.ylabel('Evaporation rate (g/h)')
plt.xticks([])
plt.xlim(0.5,1.5)
plt.ylim(2,4.5)
plt.scatter(X1,Y1,zorder=2,s=50, c="#474D6B70", linewidths=2, ec="#474D6B",label='Slow')
plt.text(0.55, 4, '$\it{Q}_{2}: 2.67$', ha='left', va='bottom',color="#474D6B",fontsize=15)
plt.text(0.55, 3.7, '$\it{Q}_{3}-\it{Q}_{1}: 0.29$', ha='left', va='bottom',color="#474D6B",fontsize=15)
# plt.yticks(np.arange(0,190,20))

figsavepath = "../Plots"
plt.savefig(f'{figsavepath}/FigureS5.svg', dpi=1200, bbox_inches='tight', transparent=True)
```
